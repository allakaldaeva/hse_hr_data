```python
import numpy as np
print("Mean = ",np.mean(df_hr.MonthlyIncome))
print("Median = ",np.median(df_hr.MonthlyIncome))
print("Dispersion = ",np.var(df_hr.MonthlyIncome))
print("Standard deviation = ",np.std(df_hr.MonthlyIncome))
```

Поиск аномальных значений

```python
# в данном случае рассматривать +3 стандартных отклонения не стоит, поскольку в таком случае аномалий не наблюдается
df_hr['anomally_sal'] = np.where(df_hr['MonthlyIncome'] > (np.mean(df_hr.MonthlyIncome) + 2.8*np.std(df_hr.MonthlyIncome)), 1, 0)
```

```python
df_anomally_sal = df_hr[df_hr['anomally_sal'] == 1]

df_anomally_sal
```
